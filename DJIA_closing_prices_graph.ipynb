# <center> Dow Jones Industrial Average (DJIA) Adjusted Closing Prices <center/> 

Here, I plot the DJIA adjusted closing prices over the period of 08-08-2008 and 07-01-2016. This plot is used in the main notebook. 

```python
# Necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style as style
import matplotlib
```

```python
# Read the data into a dataframe
stock = pd.read_csv('Data/upload_DJIA_table.csv', usecols=['Date', 'Adj Close'])
```

```python
# Check the head of the dataframe
stock.head()
```

```python
# Set the index to be the "Date" column
stock.set_index('Date', inplace=True)
```

```python
# Reverse the dataframe
stock = stock.iloc[::-1]
```

```python
# Compute moving averages for 100 days
mavg = stock['Adj Close'].rolling(window=100).mean()
```

```python
# Plot the figure
fig1 = plt.gcf()
sns.set_context('talk', font_scale=0.8)
style.use('ggplot')

ax = stock['Adj Close'].plot(figsize=(16,8), color='royalblue', label='DJIA')
ax = mavg.plot(label='moving averages = 100 days', color='k')
_ = plt.legend(loc='lower right')
_ = plt.title('Dow Jones Industrial Average (DJIA) Adjusted Closing Prices', fontsize=20, fontweight='bold')
_ = ax.yaxis.set_major_formatter(matplotlib.ticker.StrMethodFormatter('{x:,.0f}'))

plt.show()
```

```python
# Save the figure as a .png file
fig1.savefig('DJIA_closing_prices.png', dpi=100, bbox_inches='tight')
```
